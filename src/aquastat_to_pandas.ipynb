### Rename the Existing Columns

```python
aquastat_df.drop()
```

```python
aquastat_df = aquastat_df.rename(columns={
    'Unnamed: 0'  : 'country',
    'Unnamed: 1'  : 'variable_name',
    'Unnamed: 2'  : 'bin1_years_93_97',
    '1993-1997'   : 'bin1_value',
    'Unnamed: 4'  : 'availability1',
    'Unnamed: 6'  : 'bin2_years_98_02',
    '1998-2002'   : 'bin2_value',
    'Unnamed: 8'  : 'availability2',
    'Unnamed: 10' : 'bin3_years_03_07',
    '2003-2007'   : 'bin3_value',
    'Unnamed: 12' : 'availiability3',
    'Unnamed: 14' : 'bin4_years_08_12',
    '2008-2012'   : 'bin4_value',
    'Unnamed: 16' : 'availability4',
    'Unnamed: 18' : 'bin5_years_13_17',
    '2013-2017'   : 'bin5_value',
    'Unnamed: 20' : 'availability5'
})

aquastat_df.head()
```

### Remove the blank and unformatted rows

```python
aquastat_df = aquastat_df[aquastat_df.country.notnull()]
```
