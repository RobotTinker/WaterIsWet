## Get the countries with the max change in the hdi, gii and gdp per capita

```python
query_statement = "select country, `mid year`, `year bucket`, hdi, \
                    round(((urban_pop/total_pop)*100), 2) urbanized \
                    from Data \
                    where hdi is not null and gdp_per_cap is not null and gii is not null \
                    order by country"
hdi_df = pd.read_sql_query(query_statement, session.bind)
hdi_df.head(10)
```

```python
hdi_maxcount_df = hdi_df.groupby("country", as_index=False).count().sort_values("hdi", ascending=False)
hdi_maxcount_df.head()
```

```python
# get a list of countries with the most number of hdi values (We know 3 is the max)
hdi_maxcount_df = hdi_maxcount_df[hdi_maxcount_df["hdi"]==3][["country"]]
hdi_maxcount_df.count()
```

```python
hdi_maxcount_df = pd.merge(hdi_df, hdi_maxcount_df, on="country")
```

```python
# get the change in hdi for these countries
hdi_group = hdi_maxcount_df.groupby("country", as_index=False)
hdi_avg_df = hdi_group.mean()
hdi_max_df = hdi_group.max()[["country", "hdi"]]
hdi_min_df = hdi_group.min()[["country", "hdi"]]
hdi_diff_df = pd.merge(hdi_max_df, hdi_min_df, on="country")
hdi_diff_df["diff"] = hdi_diff_df["hdi_x"] - hdi_diff_df["hdi_y"]
hdi_diff_df = hdi_diff_df.sort_values("diff", ascending=False)

# get the ones with a change of >=0.1
hdi_diff_df = hdi_diff_df[hdi_diff_df["diff"]>=0.1]
hdi_diff_df
```

```python
hdi_diff_df = hdi_diff_df[["country"]]
hdi_diff_df
```

```python
# reduce the original hdi_df to only the above countries
hdi_df = pd.merge(hdi_df, hdi_diff_df, on="country")
hdi_df
```

```python
# group by year and get the avg hdi
hdi_mean_df = hdi_df.groupby("mid year").mean()
hdi_mean_df
```

```python
query_statement = "select country, `mid year`, `year bucket`, gii, \
                    round(((urban_pop/total_pop)*100), 2) urbanized \
                    from Data \
                    where hdi is not null and gdp_per_cap is not null and gii is not null \
                    order by country"
gii_df = pd.read_sql_query(query_statement, session.bind)
gii_df["mid year"].unique()
```

```python
gii_maxcount_df = gii_df.groupby("country", as_index=False).count().sort_values("gii", ascending=False)

# get a list of countries with the most number of gii values (We know 3 is the max)
gii_maxcount_df = gii_maxcount_df[gii_maxcount_df["gii"]==3][["country"]]
gii_maxcount_df = pd.merge(gii_df, gii_maxcount_df, on="country")

# get the change in gii for these countries
gii_group = gii_maxcount_df.groupby("country", as_index=False)
gii_avg_df = gii_group.mean()
gii_max_df = gii_group.max()[["country", "gii"]]
gii_min_df = gii_group.min()[["country", "gii"]]
gii_diff_df = pd.merge(gii_max_df, gii_min_df, on="country")
gii_diff_df["diff"] = gii_diff_df["gii_x"] - gii_diff_df["gii_y"]
gii_diff_df = gii_diff_df.sort_values("diff", ascending=False)

# get the ones with a change of >=0.1
gii_diff_df = gii_diff_df[gii_diff_df["diff"]>=0.1]

gii_diff_df = gii_diff_df[["country"]]

# reduce the original gii_df to only the above countries
gii_df = pd.merge(gii_df, gii_diff_df, on="country")

# group by year and get the avg gii
gii_mean_df = gii_df.groupby("mid year").mean()
gii_mean_df
```

```python
query_statement = "select country, `mid year`, `year bucket`, gdp_per_cap, \
                    round(((urban_pop/total_pop)*100), 2) urbanized \
                    from Data \
                    where hdi is not null and gdp_per_cap is not null and gii is not null \
                    order by country"
gdp_df = pd.read_sql_query(query_statement, session.bind)
gdp_df["mid year"].unique()
```

```python
gdp_maxcount_df = gdp_df.groupby("country", as_index=False).count().sort_values("gdp_per_cap", ascending=False)

# get a list of countries with the most number of gdp values (We know 3 is the max)
gdp_maxcount_df = gdp_maxcount_df[gdp_maxcount_df["gdp_per_cap"]==3][["country"]]
gdp_maxcount_df = pd.merge(gdp_df, gdp_maxcount_df, on="country")

# get the change in gdp for these countries
gdp_group = gdp_maxcount_df.groupby("country", as_index=False)
gdp_avg_df = gdp_group.mean()
gdp_max_df = gdp_group.max()[["country", "gdp_per_cap"]]
gdp_min_df = gdp_group.min()[["country", "gdp_per_cap"]]
gdp_diff_df = pd.merge(gdp_max_df, gdp_min_df, on="country")
gdp_diff_df["diff"] = gdp_diff_df["gdp_per_cap_x"] - gdp_diff_df["gdp_per_cap_y"]
gdp_diff_df = gdp_diff_df.sort_values("diff", ascending=False)

# get the ones with a change of >=0.1
gdp_diff_df = gdp_diff_df[gdp_diff_df["diff"]>=0.1]

gdp_diff_df = gdp_diff_df[["country"]]

# reduce the original gdp_df to only the above countries
gdp_df = pd.merge(gdp_df, gdp_diff_df, on="country")

# group by year and get the avg gdp
gdp_mean_df = gdp_df.groupby("mid year").mean()
gdp_mean_df
```
